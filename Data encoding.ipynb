```python
time_info = data['time']
average_time = []
for st in time_info:
    try:
        s, k = 0, 0
        sr = 0
        arr = st.split(' ')
        for i, item in enumerate(arr, start=1):
            if (i%2)!=0:
                s+=float(item)
                k+=1
        sr = s/k
    except:
        sr = 0
        
    average_time.append(sr)
data.drop('time', axis = 1, inplace = True)
data['average_time'] = average_time    

```

```python
toilet_info = data['toilet']
toilet = []
for st in toilet_info:
    try:
        s = 0
        arr = st.split(' ')
        for i, item in enumerate(arr, start=1):
            if (i%2)!=0:
                s+=float(item)
    except:
        s = 0
        
    toilet.append(s)
data.drop('toilet', axis = 1, inplace = True)
data['toilet'] = toilet    

```

```python
balcony_info = data['balcony']
balcony = []
for st in balcony_info:
    try:
        s = 0
        arr = st.split(' ')
        for i, item in enumerate(arr, start=1):
            if (i%2)!=0:
                s+=float(item)
    except:
        s = 0
        
    balcony.append(s)
data.drop('balcony', axis = 1, inplace = True)
data['balcony'] = balcony 

```

```python
service_lift = []
elevators_new = []
elevators = data['elevators']

for st in elevators:
    try:
        arr = st.split(' ')
        f=0
        for i, item in enumerate(arr, start=1):
            if i==1:
                elevators_new.append(int(item))
            if i==3:
                f=1
                service_lift.append(int(item))
        if f == 0 :
            service_lift.append(0)
    except:
        elevators_new.append(0)
        service_lift.append(0)

data.drop('elevators', axis = 1, inplace = True)
data['elevators'] = elevators_new    
data['service lift'] = service_lift  
```

<h3>Filling missing values</h3>

```python
# filling missing values

data = data.dropna (subset=['price'])

data['type'] = data['type'].fillna(data['type'].mode()[0])

data['toilet'] = data['toilet'].fillna(1)
data['toilet'].replace(0, 1, inplace=True)

data['mortgage'] = data['mortgage'].fillna('No')

data['parking'] = data['parking'].fillna('No')

data.sort_values(by=['type', 'meters', 'price'], ascending=[True, True, True], inplace=True)
data['living_area'] = data['living_area'].fillna(method='ffill')

data.sort_values(by=['district', 'floor 2'], ascending=[True, True], inplace=True)
data['year_at_home'] = data['year_at_home'].fillna(method='ffill')

data.sort_values(by=['rooms', 'district', 'price'], ascending=[True, True, True], inplace=True)
data['window_view'] = data['window_view'].fillna(method='ffill')

data['repair '] = data['repair '].fillna('Без ремонта')

data.sort_values(by=['district', 'price'], ascending=[True, True], inplace=True)
data['type_of_house'] = data['type_of_house'].fillna(method='ffill')


data['floor 1'] = data['floor 1'].fillna(data['floor 1'].mode()[0])
data['floor 2'] = data['floor 2'].fillna(data['floor 2'].mode()[0])

data.isna().sum()
```
